```python
import heapq
from heapq import heappop, heappush
 
 
def isLeaf(foo):
    return foo.l is None and foo.r is None
 
 

class Node:
    def __init__(self, char, frequen, l=None, r=None):
        self.char = char
        self.frequen = frequen
        self.l = l
        self.r = r
 
   
    def __lt__(self, other):
        return self.frequen < other.frequen
 
 

def encode(foo, str, huffman_code):
 
    if foo is None:
        return
 
    
    if isLeaf(foo):
        huffman_code[foo.char] = str if len(str) > 0 else '1'
 
    encode(foo.l, str + '0', huffman_code)
    encode(foo.r, str + '1', huffman_code)
 


def decode(foo, index, str):
 
    if foo is None:
        return index
 
    if isLeaf(foo):
        print(foo.char, end='')
        return index
 
    index = index + 1
    foo = foo.l if str[index] == '0' else foo.r
    return decode(foo, index, str)

 

def buildHuffmanTree(string):
 
    
    if len(string) == 0:
        return
 
    
    frequen = {i: string.count(i) for i in set(string)}
 
    
    pri = [Node(k, v) for k, v in frequen.items()]
    heapq.heapify(pri)
 
    
    while len(pri) != 1:
 
        
 
        l = heappop(pri)
        r = heappop(pri)
 
        
 
        total = l.frequen + r.frequen
        heappush(pri, Node(None, total, l, r))
 
    
    foo = pri[0]
 
    
    huffmanCode = {}
    encode(foo, "", huffmanCode)

 
    
    str = ""
    for c in string:
        str += huffmanCode.get(c)
     
    file = open('text2.txt', 'w')
    file.write(str)
    file.close()
    print("The decoded string is:", end=' ')
   
 
    if isLeaf(foo):
        
        while foo.frequen > 0:
            print(foo.char, end='')
            foo.frequen = foo.frequen - 1
    else:
        index = -1
        while index < len(str) - 1:
            index = decode(foo, index, str)
 
 

if __name__ == '__main__':
 
    buildHuffmanTree(string)
```
